#### 自动求导

![1763954661264](../note/image/machine-learning/1763954661264.png)


```python
import torch
x = torch.arange(4.0)
x  #创建一个x=。。。。长为4的向量x
```

![1763954864598](../note/image/machine-learning/1763954864598.png)

```python
x.requires_grad_(True) # 等价于 `x = torch.arange(4.0,requires_grad=True)`  # 告诉需要梯度
x.grad # 默认值是None  # 对于x的梯度存储在这第

# 计算y

y = 2* torch.dot(x,x)  # 计算内积*2
y  # 一列乘一行计算出的是一个标量
```
