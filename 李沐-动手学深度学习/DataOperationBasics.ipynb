```python
# 数据操作
import torch
```

```python
# 张量表示一个数值组成的数组，这个数组可能有多个维度
x = torch.arange(12)
x
```

```python
# 通过张量的shape属性来访问张量的姓皇族那个和张量重元素的总数
# 输出的是一个 torch.Size 对象。
x.shape
```

```python
# 输返回张量中元素的总数，永远是一个标量
x.numel()
```

```python
# 改变一个张量的形状而不改变元素数量和元素值
# reshape的形状所能容纳的数据数量必须和原数据的数量相等
x = x.reshape(3,4)
x
```

```python
# 创建全0元素
torch.zeros((2, 3, 4))
```

```python
# 创建全1元素
torch.ones((2, 3, 4))
```

```python
# 创建张量
torch.tensor(
    [
        [2, 1, 4, 3],
        [1, 2, 3, 4],
        [4, 3, 2, 1],
    ]
)
```

```python
#  带小数点自动识别为浮点数
x = torch.tensor(
    [1.0, 2, 4, 8],
)
y = torch.tensor(
    [2, 2, 2, 2],
)
# 可以进行相关运算，每个子元素进行计算
x + y, x - y, x * y, x / y, x**y
```

```python
# exp(x)=(ex1​,ex2​,...,exn​)
# 对一个张量的每一元素计算e的x次方
torch.exp(x)
```

```python
# 把多个张量结合再一起
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))

Y = torch.tensor(
    [
        [2.0, 1, 4, 3],
        [1, 2, 3, 4],
        [4, 3, 2, 1],
    ]
)
# 合并两个张量
torch.cat(
    (X, Y),
    dim=0,  # 在第0维合并
),
```

```python
torch.cat(
    (X, Y),
    dim=1,  # 在第一维合并
)
```

```python
# 使用逻辑运算符构建二维张量
X == Y
```

```python
# 对张量中的所有元素进行求和会产生一个只有一个元素的张量
X.sum()
# 一定注意生成的是一个张量
```

```python
# 即使形状不同，我们仍然可以通过 广播机制来执行元素操作
# 注意！ 广播机制可以发生在 形状不同的张量之间，
# 但是无法发生在维度不同的张量之间
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a,b
```

```python
# 底层运算操作时
# 会把同一纬度下的 小数向大数靠拢
a+b
```

```python
# 元素可以切片访问
X,X[-1], X[0:3]
```

```python
# 对单个元素进行赋值
X[1,2] = 9
X
```

```python
# 对多个元素进行统一赋值
# 按区域赋值
X[0:2,:] = 12
X
```

```python
# 运行一些操作可能会导致为新结果分配内存
before = id(Y)
Y = Y + X  # 因为发生了重复使用，原内存已经被析构，导致内存重新分配
id(Y) == before
```

```python
# 执行原地操作
z = torch.zeros_like(Y)
print("id(Z):", id(z))
z[:] = X + Y
print("id(Z):", id(z))
```
